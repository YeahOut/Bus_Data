```python
import pandas as pd
import numpy as np

# haversine 함수 정의
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    m = km * 1000  # convert km to m
    return m

# 데이터 로드
population_data = pd.read_csv('수원시_시간대별_유동인구(2020).csv')

# 데이터 병합을 위한 새로운 DataFrame 생성
merged_data = bus_data.copy()

# 유동인구 데이터의 각 row에 대해 반복
for index, pop_row in population_data.iterrows():
    # 버스 정류장 데이터의 각 row에 대해 반복
    for j, bus_row in bus_data.iterrows():
        # haversine 함수를 사용하여 거리 계산
        distance = haversine(bus_row['lon'], bus_row['lat'], pop_row['lon'], pop_row['lat'])

        # 거리가 100m 이내일 경우 유동인구 데이터를 병합
        if distance <= 100:
            for col in population_data.columns:
                if col not in ['lon', 'lat']:
                    # 유동인구 데이터의 column 이름을 변경하여 병합
                    new_col_name = f"{col}_pop"
                    if new_col_name not in merged_data.columns:
                        merged_data[new_col_name] = None
                    merged_data.at[j, new_col_name] = pop_row[col]

# 결과 출력
print(merged_data.head())
```
